```python
key_sc = 'b5bdfa407799caff69527eda8144ae1c'
from pyscopus import Scopus
sc = Scopus(key_sc)
```

```python
search_df = sc.search("KEY(interdisciplinary collaboration AND negotiation OR policy)", count=20,view='COMPLETE')
```

```python
search_df.head(2)
```

```python
x = list()
for id in search_df['scopus_id']:
    try:
        x.append(sc.retrieve_abstract(id))
    except:
        None

import pandas as pd
x = pd.DataFrame(x)
```

```python
x.head(2)
```

```python
pd.merge(search_df,x,how='left')
```
